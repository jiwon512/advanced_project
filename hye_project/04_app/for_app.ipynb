```python
def explore_pairs(df: pd.DataFrame,
                  pairs: list[tuple[str, str]]) -> dict:
    """
    • pairs에 ('cat_col', 'related_col') 튜플 리스트를 넘기면,
      각 튜플마다 cat_col의 각 카테고리별로 related_col의 요소 리스트를 반환합니다.
    • 반환형: { (cat_col, related_col): { category_value: [elements…], … }, … }
    """
    result: dict[tuple[str, str], dict] = {}
    for cat_col, related_col in pairs:
        mapping: dict = {}
        categories = df[cat_col].dropna().unique()
        for cat in categories:
            elems = (
                df.loc[df[cat_col] == cat, related_col]
                  .dropna()
                  .unique()
                  .tolist()
            )
            mapping[cat] = elems
        result[(cat_col, related_col)] = mapping
    return result

'''
# 사용 예시
pairs = [
    ('room_type', 'neighbourhood'),
    ('neighbourhood_group_cleansed', 'room_type'),
    # 필요에 따라 더 추가
]
res = explore_pairs(df, pairs)

# 보기 좋게 출력
for (cat_col, rel_col), mapping in res.items():
    print(f"=== ({cat_col}, {rel_col}) ===")
    for cat, elems in mapping.items():
        print(f"{cat!r} → {len(elems)}개: {elems}")
    print()

'''
```

```python
pairs = [
    ('neigh_cluster_reduced', 'neighbourhood_cleansed'),
    ('room_type_ord', 'room_type'),
    ('room_new_type_ord', 'room_new_type')]
res = explore_pairs(df, pairs)

# 보기 좋게 출력
for (cat_col, rel_col), mapping in res.items():
    print(f"=== ({cat_col}, {rel_col}) ===")
    for cat, elems in mapping.items():
        print(f"{cat!r} → {len(elems)}개: {elems}")
    print()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
import joblib

# 1) 기존에 만든 preprocessor, stack 모델 불러오기
preprocessor = ColumnTransformer([
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), cat_cols)
], remainder='passthrough')

stack = joblib.load('/Users/hyeom/Documents/GitHub/advanced_project/hye_project/03_MachineLearning/final_ensemble_model_with_full_metrics.pkl')

# 2) 파이프라인 생성
pipeline = Pipeline([
    ('pre', preprocessor),
    ('model', stack)
])

# 3) 파이프라인 저장
joblib.dump(pipeline, 'price_prediction_pipeline.pkl')
```
