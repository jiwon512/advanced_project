### booking_info

```python
booking_info = df[['id', 'host_id', 'price', 'estimated_occupancy_l365d', 'estimated_revenue_l365d', 'availability_60', 'availability_365', 'minimum_nights', 'maximum_nights', 'instant_bookable']]
booking_info

# estimated occupancy and revenue 가 0 이 많아서 데이터 살펴보다보니 마지막 후기 옛날인 경우들도 존재 운휴중일 가능성 존재 -- 이건 어떻게?
# estimated occupancy 0 인 값을 0 제외?
```

```python
booking_info['price'] = booking_info['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
booking_info.isnull().sum()
```

```python
booking_info['instant_bookable'] =  booking_info['instant_bookable'].map({'f':0, 't':1})
```

```python
booking_info['minimum_nights'].unique()
```

```python
booking_info['maximum_nights'].unique()
```

```python
booking_info['is_long_term'] = (booking_info['minimum_nights'] >= 28).astype(int)
```

```python
booking_info
```

```python
# estimated_occupancy_l365d, estimated_revenue_l365d 0인 비율 구하기

zero_occupancy_pct = (booking_info['estimated_occupancy_l365d'] == 0).sum()/len(booking_info)
zero_revenue_pct = (booking_info['estimated_revenue_l365d'] == 0).sum()/len(booking_info)

print(f'이전 사용 0인 비율: {zero_occupancy_pct:.2f}')
print(f'이전 수입 0인 비율: {zero_revenue_pct:.2f}')
```

```python
import pandas as pd
import ast
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df['amenities'].isnull().sum()
```

```python
amenities_info = df[['id', 'host_id', 'host_location', 'neighbourhood_cleansed', 'neighbourhood_group_cleansed', 'amenities']]
```

```python
def parse_amenities(x):
    try:
        return [a.strip().strip('"').strip("'") for a in ast.literal_eval(x)]
    except:
        return []
    
amenities_info['amenities_list'] = amenities_info['amenities'].apply(parse_amenities)
```

```python
amenities_info
```

```python
grouped_s = amenities_info.groupby('neighbourhood_cleansed')['amenities_list'].sum().reset_index()

def get_top_amenities(lst, topn=10):
    return pd.Series(lst).value_counts().head(topn).index.tolist()

grouped_s['top10_amenities'] = grouped_s['amenities_list'].apply(get_top_amenities)

grouped_s
```

```python
grouped_s.to_csv('neighbourhood_amenities_top10.csv', index=False)
```

```python
grouped_b = amenities_info.groupby('neighbourhood_group_cleansed')['amenities_list'].sum().reset_index()

def get_top_amenities(lst, topn=10):
    return pd.Series(lst).value_counts().head(topn).index.tolist()

grouped_b['top10_amenities'] = grouped_b['amenities_list'].apply(get_top_amenities)

grouped_b
```
