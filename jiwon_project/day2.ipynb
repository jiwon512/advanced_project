```python
import matplotlib.pyplot as plt
import seaborn as sns

# 수치형 변수만 추출해서 상관계수 계산
corr = df.corr(numeric_only=True)

plt.figure(figsize=(18, 16))  # 크기는 데이터 컬럼 수에 따라 조정!
sns.heatmap(
    corr,
    annot=True,          # 숫자값 표시
    fmt='.2f',           # 소수점 2자리
    cmap='coolwarm',     # 색상 (원하는 colormap으로 변경 가능)
    square=True,
    linewidths=.5,
    cbar_kws={"shrink": .75}
)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
